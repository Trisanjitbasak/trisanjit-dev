# WEATHER DATA ANALYSIS WITH PANDAS

```python
import pandas as pd
```

### 1️⃣ Load the CSV file into a pandas DataFrame and display the first 5 rows.

```python
data = pd.read_csv("/home/trisanjit/Desktop/weather_data.csv")
```

```python
data.head(5)
```

### 2️⃣ Find the total number of missing values in each column.

```python
data.isnull()
```

```python
data.isnull().sum()
```

### 3️⃣ What is the average temperature recorded in the dataset?

```python
data["Temperature (°C)"].mean()
```

### 4️⃣ Find all records where the wind speed is greater than 20 km/h.

```python
data[(data["Wind Speed (km/h)"] > 20.)]
```

### 5️⃣ How many times did it rain in the dataset?

```python
data[(data["Condition"] == "Rainy")].count()
```

### 6️⃣ Which day had the highest humidity recorded?

```python
max = data["Humidity (%)"].max()
data[(data["Humidity (%)"] == max)]
```

### 7️⃣ Find the average temperature for each weather condition (e.g., Rainy, Sunny, Cloudy, etc.).

#### Rainy

```python
data.groupby("Condition")["Temperature (°C)"].mean()
```

### 8️⃣ On which date was the highest temperature recorded?

```python
max = data["Temperature (°C)"].max()
data[(data["Temperature (°C)"] == max)]
```

### 9️⃣ Sort the dataset in descending order of temperature.

```python
data.sort_values(by="Temperature (°C)" ,ascending=False).head()
```

### 🔟 Find the top 10 days with the highest temperatures recorded.

```python
data.sort_values(by="Temperature (°C)" ,ascending=False).head(10)
```
