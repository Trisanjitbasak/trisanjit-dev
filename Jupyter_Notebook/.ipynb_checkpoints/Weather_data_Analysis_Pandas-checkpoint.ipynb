### 2Ô∏è‚É£ Find the total number of missing values in each column.

```python
data.isnull()
```

```python
data.isnull().sum()
```

### 3Ô∏è‚É£ What is the average temperature recorded in the dataset?

```python
data["Temperature (¬∞C)"].mean()
```

### 4Ô∏è‚É£ Find all records where the wind speed is greater than 20 km/h.

```python
data[(data["Wind Speed (km/h)"] > 20.)]
```

### 5Ô∏è‚É£ How many times did it rain in the dataset?

```python
data[(data["Condition"] == "Rainy")].count()
```

### 6Ô∏è‚É£ Which day had the highest humidity recorded?

```python
max = data["Humidity (%)"].max()
data[(data["Humidity (%)"] == max)]
```

### 7Ô∏è‚É£ Find the average temperature for each weather condition (e.g., Rainy, Sunny, Cloudy, etc.).

#### Rainy

```python
data.groupby("Condition")["Temperature (¬∞C)"].mean()
```

### 8Ô∏è‚É£ On which date was the highest temperature recorded?

```python
max = data["Temperature (¬∞C)"].max()
data[(data["Temperature (¬∞C)"] == max)]
```

### 9Ô∏è‚É£ Sort the dataset in descending order of temperature.

```python
data.sort_values(by="Temperature (¬∞C)" ,ascending=False).head()
```

### üîü Find the top 10 days with the highest temperatures recorded.

```python
data.sort_values(by="Temperature (¬∞C)" ,ascending=False).head(10)
```
